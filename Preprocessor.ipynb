```python
import pandas as pd

games_df = pd.read_csv('games.csv')
games_df.head()
```

```python
def columnremover(dataset, name):
    dataset=dataset.drop([name], axis=1)
    return dataset
games_df = columnremover(games_df,'About the game')
games_df = columnremover(games_df,'Screenshots')
games_df = columnremover(games_df,'Movies')
games_df = columnremover(games_df,'Notes')
games_df = columnremover(games_df,'Required age')
games_df = columnremover(games_df,'Support url')
games_df = columnremover(games_df,'Support email')
games_df = columnremover(games_df,'Header image')
games_df = columnremover(games_df,'Metacritic url')
games_df = columnremover(games_df,'Windows')
games_df = columnremover(games_df,'Mac')
games_df = columnremover(games_df,'Linux')
games_df = columnremover(games_df,'Full audio languages')
games_df = columnremover(games_df,'Supported languages')
games_df = columnremover(games_df,'Average playtime two weeks')
games_df = columnremover(games_df,'Median playtime two weeks')
games_df = columnremover(games_df,'Genres')
games_df = columnremover(games_df,'Tags')
games_df = columnremover(games_df,'Categories')
games_df = columnremover(games_df,'Website')
games_df = columnremover(games_df,'Release date')
games_df = columnremover(games_df,'Estimated owners')
games_df = columnremover(games_df,'Name')
games_df = columnremover(games_df,'Developers')
games_df = columnremover(games_df,'Publishers')
games_df = columnremover(games_df,'Reviews')
games_df = columnremover(games_df,'AppID')
games_df = columnremover(games_df,'Score rank')
games_df = columnremover(games_df,'Price')
games_df.fillna(0)
games_df.fillna(0, axis=1)
```
